```python
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def columnPlot(data,X):
    
    sns.countplot(x=X, data=data)
    plt.xlabel(str(X))
    plt.ylabel('Number of students')
    plt.title('Distribution of students by '+str(X))
    plt.show()
    return None

def create_3d_plot(data, x_col, y_col, z_col):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    ax.scatter(data[x_col], data[y_col], data[z_col], c='r', marker='o')

    ax.set_xlabel(x_col)
    ax.set_ylabel(y_col)
    ax.set_zlabel(z_col)

    plt.title(f'3D Plot of {x_col}, {y_col}, and {z_col}')
    plt.show()
    return None

def PlotBubble(data,x_col,y_col,color):
    df = pd.DataFrame(data)
    counts = df.groupby([x_col, y_col]).size().reset_index(name='count')
    fig, ax = plt.subplots()
    sns.scatterplot(x=x_col, y=y_col, size='count', sizes=(40, 500), data=counts, ax=ax, legend=False, color=color)
    plt.title("Graph of "+str(x_col)+" vs "+str(y_col))
    return None

def Plot3d(data, x_col,y_col):
    df = pd.DataFrame(data)

    counts = df.groupby([x_col, y_col]).size().reset_index(name='count')
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_trisurf(counts[x_col], counts[y_col], counts['count'], cmap='viridis', linewidth=0.2, edgecolor='k')
    plt.title('3d graph of '+str(x_col)+' vs '+str(y_col))
    plt.show()
    return None
    

path = 'student-mat.csv'
data = pd.read_csv(path,header=0)

x_column = 'Dalc'
y_column = 'G1'

Plot3d(data,x_column,y_column)

x_column = 'Walc'
y_column = 'G1'

Plot3d(data,x_column,y_column)

x_column = 'Dalc'
y_column = 'G2'

Plot3d(data,x_column,y_column)

x_column = 'Walc'
y_column = 'G2'

Plot3d(data,x_column,y_column)

x_column = 'Dalc'
y_column = 'G3'

Plot3d(data,x_column,y_column)

x_column = 'Walc'
y_column = 'G3'

Plot3d(data,x_column,y_column)

```
